문제 1. 단일표본 t-test: 배달시간 평균 검정

```python
from pathlib import Path
import pandas as pd
from scipy import stats
# CSV 경로 설정
DATA_DIR = Path('/content/drive/MyDrive/Colab Notebooks')
file_path = DATA_DIR / 'delivery_time.csv'
# CSV 불러오기
df = pd.read_csv(file_path)
# 데이터 확인
print('[데이터 미리보기]')
print(df.head(), '\n')
# 분석 대상
x = df['delivery_min'].dropna()
# 단일표본 t-test
# H0: 평균 배달시간은 30분 이상이다.
# H1: 평균 배달시간은 30분 미만이다.
result = stats.ttest_1samp(x, popmean=30, alternative='less')
mean_x = x.mean()
n = len(x)
se = stats.sem(x)
ci_low, ci_high = stats.t.interval(0.95, df=n-1, loc=mean_x, scale=se)
print('[분석 결과]')
print('표본 수:', n)
print('표본평균:', round(mean_x, 4))
print('t-statistic:', round(result.statistic, 4))
print('p-value:', round(result.pvalue, 6))
print('95% 신뢰구간:', (round(ci_low, 4), round(ci_high, 4)))
print('\n[결과 해석]')
if result.pvalue < 0.05 and mean_x < 30:
  print("p-value가 0.05보다 작고 평균이 30보다 작으므로, 평균 배달시간은 30분 미만이라고 해석할 수 있습니다.")
else:
  print('p-value가 0.05 이상이거나 평균이 30보다 작지 않으므로, 평균 배달시간이 30분 미만이라고 결론 내리기 어렵습니다.')
```

문제 2. 대응표본 t-test: 교육 전후 업무처리시간 비교

```python
# 문제 2. 대응표본 t-test: 교육 전후 업무처리시간 비교
from pathlib import Path
import pandas as pd
from scipy import stats
DATA_DIR = Path('/content/drive/MyDrive/Colab Notebooks')
file_path = DATA_DIR / 'before_after_worktime.csv'
df = pd.read_csv(file_path)
print('[데이터 미리보기]')
print(df.head(), '\n')
before = df['before_min']
after = df['after_min']
diff = before - after
# H0: 교육 전후 평균 차이는 없다.
# H1: 교육 후 업무처리시간이 감소했다.
result = stats.ttest_rel(before, after, alternative='greater')
print('[분석 결과]')
print('표본 수:', len(df))
print('평균 차이(before - after):', round(diff.mean(), 4))
print('t-statistic:', round(result.statistic, 4))
print('p-value:', round(result.pvalue, 6))
print('\n[결과 해석]')
if result.pvalue < 0.05 and diff.mean() > 0:
  print('교육 전후 차이가 유의하며, before-after 평균이 양수이므로 교육 후 업무처리시간이 감소했다고 볼 수 있습니다.')
else:
  print('교육 후 업무처리시간이 감소했다고 통계적으로 확인하기 어렵습니다.')
```
